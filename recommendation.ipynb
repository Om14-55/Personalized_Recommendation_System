# User-based **CF**

```python
user_item_matrix = df.pivot_table(
    index='user_id',
    columns='product_id',
    values='rating'
)
```

```python
# Missing values ko fill karo
user_item_filled = user_item_matrix.fillna(0)
# Convert to sparse matrix
from scipy.sparse import csr_matrix
user_item_sparse = csr_matrix(user_item_filled.values)
# Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_item_sparse)

user_sim_df = pd.DataFrame(
    user_similarity,
    index=user_item_filled.index,
    columns=user_item_filled.index
)

```

```python
def get_user_based_recommendations(user_id, user_item_matrix, user_sim_matrix, top_n=5):
    if user_id not in user_item_matrix.index:
        return f"User '{user_id}' not found in the dataset."

    sim_scores = user_sim_matrix[user_id].drop(labels=[user_id])
    top_users = sim_scores.sort_values(ascending=False).head(10)

    weighted_ratings = user_item_matrix.loc[top_users.index].T.dot(top_users)
    sim_sums = (user_item_matrix.loc[top_users.index].notna().T.dot(top_users)) + 1e-9
    scores = weighted_ratings / sim_sums

    already_rated = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id].notna()].index
    scores = scores.drop(index=already_rated, errors='ignore')

    return scores.sort_values(ascending=False).head(top_n)

# Example usage:
recommendations = get_user_based_recommendations("Ricky", user_item_matrix, user_sim_df, top_n=5)
print(recommendations)
```

# Item-based Collaborative Filtering

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Step 1: Fill NaNs with 0
user_item_matrix_filled = user_item_matrix.fillna(0)

# Step 2: Transpose for item-item similarity
item_item_matrix = cosine_similarity(user_item_matrix_filled.T)

# Step 3: Convert to DataFrame with proper labels
item_sim_df = pd.DataFrame(item_item_matrix,
                           index=user_item_matrix.columns,
                           columns=user_item_matrix.columns)
```

```python
def get_item_based_recommendations(user_id, user_item_matrix, item_sim_matrix, top_n=5):
    if user_id not in user_item_matrix.index:
        return f"User '{user_id}' not found in the dataset."

    # Get all items the user has rated
    user_ratings = user_item_matrix.loc[user_id].dropna()

    scores = pd.Series(dtype='float64')

    # For each item the user has rated
    for item, rating in user_ratings.items():
        # Get similarity scores for this item
        similar_scores = item_sim_matrix[item] * rating
        scores = scores.add(similar_scores, fill_value=0)

    # Remove already rated items
    scores = scores.drop(labels=user_ratings.index, errors='ignore')

    # Return top-N recommendations
    return scores.sort_values(ascending=False).head(top_n)
```

```python
recommendations_item_cf = get_item_based_recommendations("Ricky", user_item_matrix, item_sim_df, top_n=5)
print(recommendations_item_cf)
```

# **for svd and nmf**

```python
# 2. Create a Reader to define the rating scale
reader = Reader(rating_scale=(1, 5))  # adjust if needed

# 3. Convert pandas DataFrame to a Surprise Dataset
data = Dataset.load_from_df(df[['user_id', 'product_id', 'rating']], reader)

# 4. Split using Surprise's `train_test_split` (this returns Surprise-style data, not pandas)
train_df, test_df = train_test_split(data, test_size=0.2, random_state=42)
```

```python
from collections import defaultdict

def get_top_n(predictions, k=5):
    '''Return top-k recommendations for each user from predictions'''
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Sort and retrieve top k
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = [iid for (iid, _) in user_ratings[:k]]
    return top_n

def precision_at_k(predictions, k=5, threshold=4):
    '''Compute average Precision@K'''
    top_n = get_top_n(predictions, k)
    hits = 0
    total = 0

    # For each user
    for uid, user_ratings in top_n.items():
        # Relevant items (actual rating >= threshold)
        relevant_items = [iid for (uid2, iid, true_r, est, _) in predictions if uid2 == uid and true_r >= threshold]
        if not relevant_items:
            continue
        hit_count = len(set(user_ratings) & set(relevant_items))
        hits += hit_count
        total += k

    return hits / total if total else 0
```

```python
rmse_scores = []
mae_scores = []
precision_scores = []

# 1. User-based CF
user_cf = KNNBasic(sim_options={'name': 'cosine', 'user_based': True})
user_cf.fit(train_df)
user_preds = user_cf.test(test_df)
rmse_scores.append(accuracy.rmse(user_preds, verbose=False))
mae_scores.append(accuracy.mae(user_preds, verbose=False))
precision_scores.append(precision_at_k(user_preds, k=5))

# 2. Item-based CF
item_cf = KNNBasic(sim_options={'name': 'cosine', 'user_based': False})
item_cf.fit(train_df)
item_preds = item_cf.test(test_df)
rmse_scores.append(accuracy.rmse(item_preds, verbose=False))
mae_scores.append(accuracy.mae(item_preds, verbose=False))
precision_scores.append(precision_at_k(item_preds, k=5))
# 3. Matrix Factorization - SVD
svd = SVD()
svd.fit(train_df)
svd_preds = svd.test(test_df)
rmse_scores.append(accuracy.rmse(svd_preds, verbose=False))
mae_scores.append(accuracy.mae(svd_preds, verbose=False))
precision_scores.append(precision_at_k(svd_preds, k=5))

# 4. Matrix Factorization - NMF
nmf = NMF()
nmf.fit(train_df)
nmf_preds = nmf.test(test_df)
rmse_scores.append(accuracy.rmse(nmf_preds, verbose=False))
mae_scores.append(accuracy.mae(nmf_preds, verbose=False))
precision_scores.append(precision_at_k(nmf_preds, k=5))
```

# **Top 5 recommendation using SVD with product_id and rating using matrix factorization-based approach**

```python
# Top-N recommendation using SVD for a user
def recommend_svd_items(user_id, model, item_ids, n=5):
    predictions = [model.predict(user_id, iid) for iid in item_ids]
    sorted_predictions = sorted(predictions, key=lambda x: x.est, reverse=True)
    top_n = sorted_predictions[:n]
    return [(pred.iid, round(pred.est, 2)) for pred in top_n]

# Example usage
all_items = df["product_id"].unique()
recommendations = recommend_svd_items("Ricky", svd, all_items, n=5)
print(recommendations)
```

```python
def recommend_nmf_items(user_id, model, item_ids, n=5):
    predictions = [model.predict(user_id, iid) for iid in item_ids]
    sorted_predictions = sorted(predictions, key=lambda x: x.est, reverse=True)
    top_n = sorted_predictions[:n]
    return [(pred.iid, round(pred.est, 2)) for pred in top_n]

# Example usage
all_items = df["product_id"].unique()
recommendations = recommend_svd_items("Ricky", nmf, all_items, n=5)
print(recommendations)
```

# Build a content-based recommendation system using TF-IDF on metadata.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import pandas as pd
```

```python
df_unique = df.drop_duplicates(subset='product_id').reset_index(drop=True)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Fill missing info with empty string
df_unique['product_info'] = df_unique['product_info'].fillna('')

# Create TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_unique['product_info'])

# Cosine similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

indices = pd.Series(df_unique.index, index=df_unique['product_id'])

```

```python
def get_content_based_recommendations(product_id, cosine_sim=cosine_sim, df=df_unique, indices=indices, top_n=5):
    if product_id not in indices:
        return f"Product ID '{product_id}' not found in dataset."

    idx = indices[product_id]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]  # Skip the item itself

    product_indices = [i[0] for i in sim_scores if i[0] < len(df)]

    return df.iloc[product_indices][['product_id', 'name', 'brand', 'category_str']]

get_content_based_recommendations('B00QJDU3KY')

```

# **calculate Precision**

```python
user_product_history = df.groupby('user_id')['product_id'].apply(set)

def precision_at_k(user_id, k=5):
    # Products the user has interacted with
    relevant_products = user_product_history.get(user_id, set())
    if not relevant_products or len(relevant_products) < 2:
        return None  # Skip if no or single product

    # Pick 1 product to use as seed (simulate recommendation)
    seed_product = list(relevant_products)[0]

    # Get top-K recommendations
    try:
        recs = get_content_based_recommendations(seed_product, top_n=k)
    except:
        return None

    if isinstance(recs, str):  # In case of error message return
        return None

    # Convert recommendations to set for comparison
    recommended_products = set(recs['product_id'] if 'product_id' in recs.columns else recs)

    # Precision@K = relevant recommended / total recommended
    hits = recommended_products.intersection(relevant_products)
    return len(hits) / k

```

```python
k = 1
precisions = []

# Loop over first 100 users for evaluation
for user in user_product_history.index[:100]:
    p_at_k = precision_at_k(user, k)
    if p_at_k is not None:
        precisions.append(p_at_k)

# Final Average Precision@K
if precisions:
    avg_precision = sum(precisions) / len(precisions)
    print(f"Average Precision@{k}: {avg_precision:.4f}")
else:
    print("Not enough valid users to calculate Precision@K")
```

# **Hybrid system**

```python
def get_content_prediction(user_id, product_id):
    similar_ids = get_content_based_recommendations(product_id)
    similar_ratings = df[(df['product_id'].isin(similar_ids)) & (df['user_id'] == user_id)]['rating']
    if not similar_ratings.empty:
        return similar_ratings.mean()
    return None
```

```python
from surprise import SVD
from surprise import Dataset, Reader

# Prepare data
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id', 'product_id', 'rating']], reader)
trainset = data.build_full_trainset()
model = SVD()
model.fit(trainset)

def get_collaborative_prediction(user_id, product_id):
    try:
        pred = model.predict(user_id, product_id)
        return pred.est
    except:
        return None
```

```python
def hybrid_prediction(user_id, product_id, alpha=0.5):
    content_pred = get_content_prediction(user_id, product_id)
    collab_pred = get_collaborative_prediction(user_id, product_id)

    if content_pred and collab_pred:
        return alpha * content_pred + (1 - alpha) * collab_pred
    elif collab_pred:
        return collab_pred
    elif content_pred:
        return content_pred
    else:
        return None
```

```python
actual = []
predicted = []

for idx, row in df.iterrows():
    user = row['user_id']
    product = row['product_id']
    true_rating = row['rating']
    pred = hybrid_prediction(user, product, alpha=0.6)

    if pred:
        actual.append(true_rating)
        predicted.append(pred)

from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

rmse = sqrt(mean_squared_error(actual, predicted))
print(f"Hybrid RMSE: {rmse:.4f}")

mae=mean_absolute_error(actual, predicted)
print(f"Hybrid MAE: {mae:.4f}")
```

```python
b=rmse
c=mae
rmse_scores.append(b)
mae_scores.append(c)
```

# **Hybrid Recommendations Check**

```python
def get_hybrid_top_k_recommendations(user_id, k):
    # Get all products not rated by user
    user_rated_products = set(df[df['user_id'] == user_id]['product_id'])
    all_products = set(df['product_id'])
    candidate_products = list(all_products - user_rated_products)

    # Score each candidate using hybrid prediction
    scored = []
    for product in candidate_products:
        pred = hybrid_prediction(user_id, product, alpha=0.6)
        if pred is not None:
            scored.append((product, pred))

    # Sort and return top-K
    scored.sort(key=lambda x: x[1], reverse=True)
    top_k = [product for product, _ in scored[:k]]
    return top_k
```

# **Fine-tune recommendation**

```python
df['predicted'] = df.apply(lambda row: hybrid_prediction(row['user_id'], row['product_id']), axis=1)
df['feedback'] = df.apply(lambda row: 1 if row['rating'] >= 4 else 0, axis=1)
```

```python
best_alpha = 0
best_rmse = float('inf')

for alpha in [0.0, 0.1, 0.2, ..., 1.0]:
    predicted = []
    actual = []

    for idx, row in df.iterrows():
        pred = hybrid_prediction(row['user_id'], row['product_id'], alpha=alpha)
        if pred:
            predicted.append(pred)
            actual.append(row['rating'])

    rmse = sqrt(mean_squared_error(actual, predicted))
    if rmse < best_rmse:
        best_rmse = rmse
        best_alpha = alpha

print(f"Optimal alpha: {best_alpha}, with RMSE: {best_rmse}")
```
